```python
# En Colab, descomenta:
# !pip install -q opencv-python matplotlib scikit-image numpy

import sys
from pathlib import Path

if str(Path(".").resolve()) not in sys.path:
    sys.path.insert(0, str(Path(".").resolve()))

import taller_modelos_color as t
import matplotlib.pyplot as plt
import numpy as np
```

## 1. Carga y conversiones RGB → HSV y RGB → CIE Lab

```python
rgb = t.load_rgb_float(None)  # None = patrón sintético; o t.load_rgb_float("mi_foto.jpg")
hsv = t.rgb_to_hsv_skimage(rgb)
lab = t.rgb_to_lab_skimage(rgb)
print("RGB shape:", rgb.shape, "HSV:", hsv.shape, "Lab:", lab.shape)
```

## 2. Canales y simulaciones

```python
fig, ax = plt.subplots(2, 3, figsize=(10, 6))
ax[0, 0].imshow(t.to_uint8(rgb)); ax[0, 0].set_title("Original"); ax[0, 0].axis("off")
ax[0, 1].imshow(t.to_uint8(t.simulate_protanopia(rgb))); ax[0, 1].set_title("Protanopía"); ax[0, 1].axis("off")
ax[0, 2].imshow(t.to_uint8(t.simulate_deuteranopia(rgb))); ax[0, 2].set_title("Deuteranopía"); ax[0, 2].axis("off")
ax[1, 0].imshow(t.to_uint8(t.low_light(rgb))); ax[1, 0].set_title("Baja luz"); ax[1, 0].axis("off")
ax[1, 1].imshow(t.to_uint8(t.color_temperature(rgb, 40))); ax[1, 1].set_title("Cálido"); ax[1, 1].axis("off")
ax[1, 2].imshow(t.to_uint8(t.monochrome(rgb))); ax[1, 2].set_title("Monocromo"); ax[1, 2].axis("off")
plt.tight_layout()
plt.show()
```

## 3. Bonus: `apply_simulation(rgb, modo)`

```python
for modo in ["invert", "cool", "lab_l"]:
    out = t.apply_simulation(rgb, modo)
    plt.figure(figsize=(4, 3))
    plt.imshow(t.to_uint8(out))
    plt.title(modo)
    plt.axis("off")
    plt.show()
```
